# Data Cleaning 2 and models

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import cross_val_score
import sklearn
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
# Read the csv with the train data with dummies

train_dummy=pd.read_csv("../input/train_dummy.csv")
```

```python
# Read the csv with the predict data with dummies

predict_dummy=pd.read_csv("../input/predict_dummy.csv")
predict=pd.read_csv("../input/predict.csv", index_col="id")
```

```python
predict_dummy
```

## Cleaning data 2

```python
# I take the dataset with the dummies and only drop column x
```

```python
train_dropped_x=train_dummy.drop(["x"], axis=1)
train_dropped_x
```

#  Models

## Linear Regression

```python
def check_model(X_train, y_train, model, X_test, y_test):
    model.fit(X_train,y_train)
    y_pred_train=model.predict(X_train)
    y_pred_test=model.predict(X_test)
    
    r2_sc_train=r2_score(y_train, y_pred_train)
    r2_sc_test=r2_score(y_test, y_pred_test)
    rmse_train=(mean_squared_error(y_train, y_pred_train))**0.5
    rmse_test=(mean_squared_error(y_test, y_pred_test))**0.5
    
    return f"r2_score of the train data: {r2_sc_train}, r2_score of the test data: {r2_sc_test}, rmse of the train data: {rmse_train}, rmse of the test data: {rmse_test}"
```

```python
# Separate into train and test
# Try with 80% data train and 25% data test

y=train_dropped_x["price"]
X=train_dropped_x.drop(columns=["price"])

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
# Apply and check the Linear Regression model

lr=LinearRegression()
check_model(X_train, y_train, lr, X_test, y_test)
```

```python
# Cross validation with train
```

```python
scores = cross_val_score(lr, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores
```

The values obtained are too high

## K Neighbors Regression

```python
# Apply and check the model

neigh = KNeighborsRegressor(n_neighbors=2)
check_model(X_train, y_train, neigh, X_test, y_test)
```

```python
# Cross validation
scoresKN = cross_val_score(neigh, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scoresKN
```

The values are very high

## Random Forest

```python
# Apply and check the model

ranf = RandomForestRegressor()
check_model(X_train, y_train, ranf, X_test, y_test)
```

```python
# Cross validation
scoresRF= cross_val_score(ranf, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scoresRF
```

Tha values for rmse are similar to the one obtained with the model, so I select this as a submission

# Decision Tree

```python
# Apply and check the model

dtr = DecisionTreeRegressor()
check_model(X_train, y_train, dtr, X_test, y_test)
```

```python
# Overfitted
```

```python
# GridSearchCV to find the best parameters to the decision tree regressor
```

```python
params = {'min_samples_leaf':np.arange(20,50,5),
          'min_samples_split':np.arange(20,50,5),
          'max_depth':np.arange(3,6),
          'min_weight_fraction_leaf':np.arange(0,0.4,0.1)}
```

```python

grid = GridSearchCV(
     estimator=DecisionTreeRegressor(),
     param_grid=params,
     verbose=3,
     scoring="neg_root_mean_squared_error",
     return_train_score=True
)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_.predict(X_test)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```
