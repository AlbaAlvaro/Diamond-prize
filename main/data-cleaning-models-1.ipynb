The columns carat, x, y and z are highly correlated with the price

```python
# If colinearity is over 90 drop this columns

columns_cor=matrix.where(np.triu(np.ones(matrix.shape), k=1).astype(np.bool))
drop_columns= [col for col in columns_cor if any(columns_cor[col]>0.90)]
drop_columns
```

```python
# Columns x,y,z have a high collinearity so I drop columns y,z because column x has a higher correlation with the price

train_dropped=train_dummy.drop(["y", "z"], axis=1)
train_dropped
```

# Models

## Linear Regression model

### Linear Regression with 20% data in test

```python
# Separate into train and test
# Try with 80% data train and 20% data test

y=train_dropped["price"]
X=train_dropped.drop(columns=["price"])

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
# Function to apply and check the models

def check_model(X_train, y_train, model, X_test, y_test):
    model.fit(X_train,y_train)
    y_pred_train=model.predict(X_train)
    y_pred_test=model.predict(X_test)
    
    r2_sc_train=r2_score(y_train, y_pred_train)
    r2_sc_test=r2_score(y_test, y_pred_test)
    rmse_train=(mean_squared_error(y_train, y_pred_train))**0.5
    rmse_test=(mean_squared_error(y_test, y_pred_test))**0.5
    
    return f"r2_score of the train data: {r2_sc_train}, r2_score of the test data: {r2_sc_test}, rmse of the train data: {rmse_train}, rmse of the test data: {rmse_test}"
```

```python
lr=LinearRegression()
check_model(X_train, y_train, lr, X_test, y_test)
```

```python
# Cross validation with train
```

```python
# Choose the rmse for the scoring

sklearn.metrics.SCORERS.keys()
```

```python
scores = cross_val_score(lr, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores
```

The scores obtained in the cross validation are very similar to the one obtain in the model, which means that the score obtained in teh model was not random.
